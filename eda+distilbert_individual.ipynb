```python
%load_ext autoreload
%autoreload 2
from IPython.display import display, Markdown

```

```python
from sample_train_test_data import sample_test_set, sample_train_set
from plots import analysis_plots, plot_classifier_results
from feature_classifiers import create_feature_classifier
```

```python
def eda_classifiers(feature, values_groups, markers, min_count=6):
    display(Markdown('### EDA'))
    analysis_plots(feature, values_groups, "",data_dir="data/")
    display(Markdown('### Classifiers'))
    display(Markdown('#### Testset sampling'))
    sample_test_set(feature, values_groups, min_count=min_count,
                    input_data_dir="data/", output_data_dir="shared_train_data/")
    
    results = []
    for seed in range(6):
        display(Markdown(f'#### Trainset sampling {seed}'))
        sample_train_set(feature, values_groups, shared=True, seed=seed,
                         min_count=min_count, input_data_dir="data/", output_data_dir="shared_train_data/")
        
        display(Markdown(f'#### Fit Classifier {seed}'))
        results.append(create_feature_classifier(feature, values_groups, seed=seed,
                                                 data_dir="shared_train_data/", min_count=min_count,
                                                 model_save_dir="shared_train_data/models/"))
    
    display(Markdown('#### Classifier plots'))
    plot_classifier_results(results, values_groups, markers)
    return results
```


# Gender

```python
feature = "gender"
values_groups = {"male": "male", "female": "female"}
markers = {"female" : ["v", "red"], "male" : ["^","blue"], 
           "mixed":["x", "green"]}
gender_classifier_results = eda_classifiers(feature, values_groups, markers)
```

```python
from collections import defaultdict
import numpy as np
metric = "f1-score"
metrics = defaultdict(lambda: defaultdict(lambda: []))
for classifier_result in gender_classifier_results:
    for train_gender in ["male", "female", "mixed"]:
        for test_gender in ["male", "female", "mixed"]:
            metrics[train_gender][test_gender].append(classifier_result[train_gender][test_gender]["report"]["True"][metric])
for train_gender in ["male", "female", "mixed"]:
    print(train_gender)
    for test_gender in ["male", "female", "mixed"]:
        print("     ", test_gender, sum(metrics[train_gender][test_gender]) / len(metrics[train_gender][test_gender]) )
```

```python
from collections import defaultdict
import numpy as np
metric = "f1-score"
metrics = defaultdict(lambda: defaultdict(lambda: []))
for classifier_result in gender_classifier_results:
    for train_gender in ["male", "female", "mixed"]:
        for test_gender in ["male", "female", "mixed"]:
            metrics[train_gender][test_gender].append(classifier_result[train_gender][test_gender]["report"]["macro avg"][metric])
for train_gender in ["male", "female", "mixed"]:
    print(train_gender)
    for test_gender in ["male", "female", "mixed"]:
        print("     ", test_gender, sum(metrics[train_gender][test_gender]) / len(metrics[train_gender][test_gender]) )
```
